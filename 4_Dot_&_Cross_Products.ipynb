## 1. The Dot Product ($\mathbf{a} \cdot \mathbf{b}$)
In high-dimensional spaces (like feature vectors or asset returns), the dot product is the fundamental measure of **projection** and **similarity**.

### The Three Operational Behaviors:
1.  **Alignment ($\theta < 90^\circ$):** The product is positive. The vectors "reinforce" each other.
2.  **Orthogonality ($\theta = 90^\circ$):** The product is zero. In Quant terms, these signals are **uncorrelated**.
3.  **Opposition ($\theta > 90^\circ$):** The product is negative. The vectors act as "hedges" to one another.

**Mathematical Duality:**
The dot product can be viewed as a linear transformation that maps a vector onto a 1D line (the number line).
$$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{n} a_i b_i = \|\mathbf{a}\| \|\mathbf{b}\| \cos(\theta)$$

```python
import numpy as np
import plotly.graph_objects as go

# Setup data
steps = 80
theta = np.linspace(0, 2 * np.pi, steps)
v1 = np.array([1, 0, 0])
axis_range = [-1.2, 1.2]

# Create static axis lines to act as a permanent grid
grid_lines = []
for i in [-1, 0, 1]:
    # X-axis grid
    grid_lines.append(go.Scatter3d(x=[-1.2, 1.2], y=[i, i], z=[0, 0], mode='lines',
                                   line=dict(color='rgba(150,150,150,0.3)', width=1), showlegend=False))
    # Y-axis grid
    grid_lines.append(go.Scatter3d(x=[i, i], y=[-1.2, 1.2], z=[0, 0], mode='lines',
                                   line=dict(color='rgba(150,150,150,0.3)', width=1), showlegend=False))

frames = []
for t in theta:
    v2 = np.array([np.cos(t), np.sin(t), 0])
    dot_val = np.dot(v1, v2)

    # Color logic: Green for Alignment, Red for Opposition
    current_color = '#00FF7F' if dot_val > 0.01 else ('#FF3131' if dot_val < -0.01 else 'white')

    frames.append(go.Frame(
        data=grid_lines + [
            go.Scatter3d(x=[0, v1[0]], y=[0, v1[1]], z=[0, v1[2]], mode='lines+markers',
                         line=dict(color='#00CCFF', width=12), name="Fixed A"),
            go.Scatter3d(x=[0, v2[0]], y=[0, v2[1]], z=[0, v2[2]], mode='lines+markers',
                         line=dict(color=current_color, width=12), name="Rotating B")
        ],
        name=str(t),
        layout=go.Layout(title_text=f"<b>Dot Product: {dot_val:.2f}</b>")
    ))

# Initial Figure Construction
fig = go.Figure(
    data=grid_lines + [
        go.Scatter3d(x=[0, 1], y=[0, 0], z=[0, 0], mode='lines+markers', line=dict(color='#00CCFF', width=12)),
        go.Scatter3d(x=[0, 1], y=[0, 0], z=[0, 0], mode='lines+markers', line=dict(color='#00FF7F', width=12))
    ],
    layout=go.Layout(
        template="plotly_dark",
        scene=dict(
            xaxis=dict(range=axis_range, title="X-Axis", showgrid=False),
            yaxis=dict(range=axis_range, title="Y-Axis", showgrid=False),
            zaxis=dict(range=axis_range, title="Z-Axis", showgrid=False),
            aspectmode='cube'
        ),
        margin=dict(l=0, r=0, b=0, t=60),
        updatemenus=[dict(
            type="buttons",
            # Positioning at top-left relative to the graph container
            x=0.1, y=0.9,
            xanchor="left", yanchor="top",
            showactive=False,
            buttons=[dict(
                label="▶ START SIMULATION",
                method="animate",
                args=[None, {
                    "frame": {"duration": 40, "redraw": True},
                    "fromcurrent": True,
                    "mode": "immediate",
                    "transition": {"duration": 0}
                }]
            )]
        )]
    ),
    frames=frames
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Setup data
steps = 100 # Increased steps for a "3B1B" smooth motion
theta = np.linspace(0, 2 * np.pi, steps)
v1 = np.array([1, 0, 0])
axis_range = [-1.2, 1.2]

frames = []
for t in theta:
    v2 = np.array([np.cos(t), np.sin(t), 0])
    dot_val = np.dot(v1, v2)

    # Color logic: Blue for positive, White for zero, Red for negative
    color = 'green' if dot_val > 0.05 else ('red' if dot_val < -0.05 else 'white')

    frames.append(go.Frame(
        data=[
            go.Scatter3d(x=[0, v1[0]], y=[0, v1[1]], z=[0, v1[2]], mode='lines+markers', line=dict(color='cyan', width=10)),
            go.Scatter3d(x=[0, v2[0]], y=[0, v2[1]], z=[0, v2[2]], mode='lines+markers', line=dict(color=color, width=10))
        ],
        name=str(t),
        layout=go.Layout(title_text=f"<b>Dot Product Analysis</b><br>Value: {dot_val:.2f} | Status: {'Hedge' if dot_val < 0 else 'Correlated'}")
    ))

fig = go.Figure(
    data=[
        go.Scatter3d(x=[0, 1], y=[0, 0], z=[0, 0], mode='lines+markers', name="Fixed A", line=dict(color='cyan', width=10)),
        go.Scatter3d(x=[0, 1], y=[0, 0], z=[0, 0], mode='lines+markers', name="Rotating B", line=dict(color='green', width=10))
    ],
    layout=go.Layout(
        template="plotly_dark",
        scene=dict(
            xaxis=dict(range=axis_range, autorange=False, backgroundcolor="black"),
            yaxis=dict(range=axis_range, autorange=False, backgroundcolor="black"),
            zaxis=dict(range=axis_range, autorange=False, backgroundcolor="black"),
            aspectmode='cube'
        ),
        updatemenus=[dict(
            type="buttons",
            x=0.1, y=0, # Position the button nicely
            buttons=[dict(
                label="▶ Start Vector Rotation",
                method="animate",
                args=[None, {"frame": {"duration": 30, "redraw": True}, "fromcurrent": True, "transition": {"duration": 0}}])
            ]
        )]
    ),
    frames=frames
)

fig.show()
```

## 2. The Cross Product ($\mathbf{a} \times \mathbf{b}$)
While the dot product is about "shared direction," the cross product is about **perpendicularity** and **area**.

### The Significance:
* **Area as Magnitude:** $\|\mathbf{a} \times \mathbf{b}\|$ is the area of the parallelogram formed by the two vectors.
* **Chirality (Right-Hand Rule):** It defines a unique direction in 3D space. If you are coding a physics engine or a 3D renderer, this is how you find the "Normal" of a surface.
* **The Determinant Link:** In 3D, the cross product is the expansion of the formal determinant:
$$\mathbf{a} \times \mathbf{b} = \det \begin{bmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ a_x & a_y & a_z \\ b_x & b_y & b_z \end{bmatrix}$$

```python
import plotly.figure_factory as ff

v1 = np.array([1, 0.5, 0])
v2 = np.array([0, 1, 0.5])
v3 = np.cross(v1, v2)

fig = go.Figure()

# Helper function to add arrows
def add_vector(vec, name, color):
    fig.add_trace(go.Scatter3d(x=[0, vec[0]], y=[0, vec[1]], z=[0, vec[2]],
                               mode='lines+markers', name=name,
                               line=dict(color=color, width=10)))

add_vector(v1, "Vector A", "cyan")
add_vector(v2, "Vector B", "magenta")
add_vector(v3, "A x B (Normal)", "yellow")

# Add a semi-transparent plane to show the span of A and B
fig.add_trace(go.Mesh3d(x=[0, v1[0], v1[0]+v2[0], v2[0]],
                        y=[0, v1[1], v1[1]+v2[1], v2[1]],
                        z=[0, v1[2], v1[2]+v2[2], v2[2]],
                        color='gray', opacity=0.3, name="Span Area"))

fig.update_layout(template="plotly_dark", title="Cross Product: Generating the Normal Vector")
fig.show()
```

## Summary Table: Dot vs. Cross Product

| Feature | Dot Product ($\mathbf{a} \cdot \mathbf{b}$) | Cross Product ($\mathbf{a} \times \mathbf{b}$) |
| :--- | :--- | :--- |
| **Output Type** | Scalar (Number) | Vector (Direction) |
| **Measures** | Projection / Similarity | Area / Perpendicularity |
| **Max Value** | When vectors are Parallel | When vectors are Orthogonal |
| **Commutative?** | Yes ($\mathbf{a}\cdot\mathbf{b} = \mathbf{b}\cdot\mathbf{a}$) | No ($\mathbf{a}\times\mathbf{b} = -(\mathbf{b}\times\mathbf{a})$) |
