# **Linear Transformation**

A matrix is not just a "box of numbers," but a linear operator T:V→W that preserves the operations of vector addition and scalar multiplication.

### **1. Defining the Linear Operator T(v)**
A transformation T is linear if it satisfies the principle of superposition:

**Additivity:** T(u+v)=T(u)+T(v)

**Homogeneity:** T(cv)=cT(v)

Instead of a static plot, we will animate the displacement of a vector v from the domain to the codomain.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 1. Configuration of the Linear Transformation
A = np.array([[2, 1],
              [-1, 1]])  # The Linear Operator
v_start = np.array([1, 2])
v_end = A @ v_start

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-2, 6); ax.set_ylim(-2, 6)
ax.set_aspect('equal')
ax.axhline(0, color='black', lw=1.5); ax.axvline(0, color='black', lw=1.5)

# 3. Reference Grid (Static)
grid_range = np.arange(-5, 10)
for i in grid_range:
    ax.axhline(i, color='gray', lw=0.5, alpha=0.2)
    ax.axvline(i, color='gray', lw=0.5, alpha=0.2)

# 4. Elements to Animate
# The Vector
vec_arrow = ax.quiver(0, 0, v_start[0], v_start[1], angles='xy', scale_units='xy',
                      scale=1, color='#4285F4', width=0.015)

# Coordinate Labels
coord_text = ax.text(v_start[0], v_start[1] + 0.2, f'({v_start[0]}, {v_start[1]})',
                     fontsize=12, fontweight='bold', color='#4285F4')

# Title for technical context
title = ax.set_title(f"Linear Mapping $T(\mathbf{{v}})$ where $\mathbf{{v}} = [{v_start[0]}, {v_start[1]}]^T$", pad=20)

# 5. Update Function
def update(frame):
    # Interpolation factor (0 to 1)
    t = frame

    # Current vector state
    v_curr = v_start + (v_end - v_start) * t

    # Update Arrow
    vec_arrow.set_UVC(v_curr[0], v_curr[1])

    # Update Text position and string
    coord_text.set_position((v_curr[0], v_curr[1] + 0.2))
    coord_text.set_text(f'({v_curr[0]:.2f}, {v_curr[1]:.2f})')

    # Color shift from Blue (start) to Google Green (end)
    # Just to add a bit of visual flair for the application
    return vec_arrow, coord_text

# Create Animation
ani = FuncAnimation(fig, update, frames=np.linspace(0, 1, 60), interval=40, blit=False)

plt.close()
HTML(ani.to_jshtml())
```

## 2. The Geometry of the Linear Operator $T$

A matrix is often taught as a static "list of numbers," but in a vector space, it is a **set of instructions** for the basis vectors.

### **The "Wind Gust" Analogy**
Imagine a drone positioned at coordinate $\mathbf{v} = [1, 1]^T$. A wind gust hits the environment, warping the coordinate system. To calculate exactly where the drone ends up, we don't need a complex simulation of every point in space. We only need to know how the wind affects a single unit step in the $x$ direction ($\hat{i}$) and the $y$ direction ($\hat{j}$).

### **Step 1: Constructing Matrix $A$**
Instead of defining a matrix $A$ arbitrarily, let's build it from the transformation $T$:

* **Original Basis (Identity):** $\hat{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ and $\hat{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$.
* **The Transformation $T$:** Suppose the wind pushes the ground such that:
    1.  The unit $x$-step ($\hat{i}$) is stretched and tilted: $T(\hat{i}) = \begin{bmatrix} 3 \\ 1 \end{bmatrix}$
    2.  The unit $y$-step ($\hat{j}$) is knocked to the left: $T(\hat{j}) = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$

**The Matrix is simply the collection of these landing spots:**
$$A = \begin{bmatrix} \mid & \mid \\ T(\hat{i}) & T(\hat{j}) \\ \mid & \mid \end{bmatrix} = \begin{bmatrix} 3 & -1 \\ 1 & 2 \end{bmatrix}$$

## The Animation: Constructing the Map
In the following script, we visualize the transition from the **Identity state** to the **Transformed state**.

Watch how the "Drone Vector" (yellow) is purely a passenger. It is defined by the linear combination $1\hat{i} + 1\hat{j}$. As the basis vectors move, the drone moves with them, maintaining its relative orientation to the transformed grid lines.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Our 'Wind Gust' Matrix
# Col 1: Where i-hat lands. Col 2: Where j-hat lands.
A = np.array([[3, -1],
              [1, 2]])

v_input = np.array([1, 1]) # The 'Drone' position

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-4, 6); ax.set_ylim(-2, 6)
ax.set_aspect('equal')
ax.grid(True, linestyle=':', alpha=0.4)

# Plotting the Basis Vectors and the Drone Vector
i_vec, = ax.plot([], [], color='#EA4335', lw=3, label=r'Transformed $\hat{i}$')
j_vec, = ax.plot([], [], color='#34A853', lw=3, label=r'Transformed $\hat{j}$')
drone_vec, = ax.plot([], [], color='#FBBC05', lw=4, label='Drone Vector $\mathbf{v}$')
trace, = ax.plot([], [], color='#FBBC05', alpha=0.3, linestyle='--')

# Text for coordinates
info_text = ax.text(-3.5, 5, '', fontsize=10, family='monospace',
                    bbox=dict(facecolor='white', alpha=0.8))

history_x, history_y = [], []

def update(frame):
    t = frame
    # LERP the matrix from Identity to A
    M_t = np.eye(2) * (1 - t) + A * t

    # Transform basis and drone
    curr_i = M_t @ np.array([1, 0])
    curr_j = M_t @ np.array([0, 1])
    curr_v = M_t @ v_input

    # Update visual lines
    i_vec.set_data([0, curr_i[0]], [0, curr_i[1]])
    j_vec.set_data([0, curr_j[0]], [0, curr_j[1]])
    drone_vec.set_data([0, curr_v[0]], [0, curr_v[1]])

    # Update trace of the drone
    history_x.append(curr_v[0])
    history_y.append(curr_v[1])
    trace.set_data(history_x, history_y)

    # Display the math in real-time
    info_text.set_text(
        f"Basis i: [{curr_i[0]:.1f}, {curr_i[1]:.1f}]\n"
        f"Basis j: [{curr_j[0]:.1f}, {curr_j[1]:.1f}]\n"
        f"Matrix A_t:\n{np.round(M_t, 2)}"
    )

    return i_vec, j_vec, drone_vec, trace, info_text

ani = FuncAnimation(fig, update, frames=np.linspace(0, 1, 60), interval=40, blit=True)
plt.legend(loc='lower right')
plt.close()
HTML(ani.to_jshtml())
```
